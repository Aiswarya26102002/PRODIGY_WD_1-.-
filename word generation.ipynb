```python
import nltk
from nltk.corpus import words
from itertools import permutations
def is_valid_word(word):
    return word.lower() in words.words()
def generate_valid_words(characters, length):
    perms = set([''.join(p) for p in permutations(characters, length)])
    valid_words = [word for word in perms if is_valid_word(word)]
    return valid_words
characters = input("Enter a string of characters: ")
word_length = int(input("Enter the word length: "))
valid_words = generate_valid_words(characters, word_length)
print(f"Valid words of length {word_length} that can be formed from '{characters}':")
print(valid_words)
```
